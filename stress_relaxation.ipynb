```python
# 

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

column = 'Modulus'

file_path = '/Volumes/KAZU-USB/220714_Rheo/0min/0min10mg_05%100s.txt'
file_path2 = '/Volumes/KAZU-USB/220714_Rheo/1min/1min10mg_05%100s.txt'
file_path3 = '/Volumes/KAZU-USB/220714_Rheo/15min/15min10mg_05%100s.txt'
file_path4 = '/Volumes/KAZU-USB/220714_Rheo/30min/30min10mg_05%100s.txt'

df = pd.read_table(file_path, header=90)
df.drop(index=0, inplace=True)
df2 = pd.read_table(file_path2, header=90)
df2.drop(index=0, inplace=True)
df3 = pd.read_table(file_path3, header=90)
df3.drop(index=0, inplace=True)
df4 = pd.read_table(file_path4, header=90)
df4.drop(index=0, inplace=True)

fig, ax = plt.subplots()

plt.plot(df['Step time'].astype('float'), df[column].astype('float'), label='0min', marker='x', linestyle='None')
plt.plot(df2['Step time'].astype('float'), df2[column].astype('float'), label='1min', marker='.', linestyle='None')
plt.plot(df3['Step time'].astype('float'), df3[column].astype('float'), label='15min', marker='.', linestyle='None')
plt.plot(df4['Step time'].astype('float'), df4[column].astype('float'), label='30min', marker='.', linestyle='None')

ax.set_xscale('log')
ax.set_yscale('log')

plt.minorticks_on()
plt.grid(which="major", color="gray", linestyle="solid")
plt.grid(which="minor", color="lightgray", linestyle="dotted")

ax.set_xlim(0.01, 100)
ax.set_ylim(0,1000000)
ax.set_xlabel('Step time[s]')
ax.set_ylabel(column + "[Pa]")

title = column

plt.legend(bbox_to_anchor=(1, 1), loc='best', borderaxespad=1)
plt.show()

fig.savefig("%s.png"%title)
```
