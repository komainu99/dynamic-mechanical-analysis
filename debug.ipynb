```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = '/Volumes/KAZU-USB/220623_Rheo/x4stg03%10_500rads (2).txt'

df = pd.read_table(file_path, header=90)
df.drop(index=0, inplace=True)


#print(df["Storage modulus"])

fig, ax = plt.subplots()

plt.plot(df['Angular frequency'].astype('float'), df['Loss modulus'].astype('float'), label='Loss modulus', marker='x', linestyle='None')
plt.plot(df['Angular frequency'].astype('float'), df['Storage modulus'].astype('float'), label="Storage modulus", marker='v', linestyle='None')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlim(10,500)
ax.set_ylim(0,1000000)
ax.set_xlabel('Angular frequency[rad/s]')
ax.set_ylabel("Storage modulus[Pa]")
plt.title("Frequency dependence measurement")

plt.legend()

plt.show()
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = '/Volumes/KAZU-USB/220623_Rheo/x4stg03%10_500rads (2).txt'

df = pd.read_table(file_path, header=90)
df.drop(index=0, inplace=True)


#print(df["Storage modulus"])

fig, ax = plt.subplots()

plt.plot(df['Angular frequency'].astype('float'), df['Loss modulus'].astype('float'), label='Loss modulus', marker='x', linestyle='None')
plt.plot(df['Angular frequency'].astype('float'), df['Storage modulus'].astype('float'), label="Storage modulus", marker='v', linestyle='None')

ax.set_yscale('log')

ax.set_xlim(10,550)
ax.set_ylim(0,1000000)
ax.set_xlabel('Angular frequency[rad/s]')
ax.set_ylabel("G'/G''[Pa]")
plt.title("Frequency dependence measurement")

plt.legend(bbox_to_anchor=(1, 0), loc='lower right', borderaxespad=1)

plt.show()
```
