```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

file_path = '/Volumes/KAZU-USB/220623_Rheo/x1stg03%10_500rads.txt'

df = pd.read_table(file_path, header=90)
df.drop(index=0, inplace=True)

#print(df)

fig, ax = plt.subplots()

plt.plot(df['Angular frequency'], df['Loss modulus'], label='Loss modulus', marker='x')
plt.plot(df['Angular frequency'], df['Storage modulus'], label="Storage modulus", marker='v')

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Angular frequency')
ax.set_ylabel("Storage modulus")
plt.title("Frequency dependence measurement")

plt.legend()

plt.show()
```
