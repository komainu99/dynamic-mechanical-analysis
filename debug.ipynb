```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

data = np.genfromtxt("/Users/yazuhidekazu/Downloads/03%_10to500rads.csv", delimiter=',')

# Frequency dependence measurement
# 横軸はdata[:,3](Angular frequency)

plt.xscale('log')
plt.yscale('log')

plt.plot(data[:,3], data[:,0], label="Storage modulus")
plt.plot(data[:,3], data[:,1], label="Loss modulus")
# plt.plot(data[:,0], data[:,2], label="Tan(delta)")
# plt.plot(data[:,0], data[:,9], label="Complex viscosity")
 
plt.title("Frequency dependence measurement")
plt.xlabel("Frequency[rad/s]")
plt.ylabel("G'/G''[Pa]")

plt.legend()
 
plt.savefig("output-figs/test220622.png")
```
