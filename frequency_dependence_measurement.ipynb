```python
# 周波数依存性

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#column = 'Storage modulus'
#column = 'Loss modulus'
column = 'Tan(delta)'

file_path = '/Volumes/Personal/Student/g16_Yazu/TA-DHR2/220623_Rheo/2_5mgml_03%_10to500rads.csv'
file_path2 = '/Volumes/Personal/Student/g16_Yazu/TA-DHR2/220623_Rheo/6_25mgml_03%_10to500rads.csv'
file_path3 = '/Volumes/Personal/Student/g16_Yazu/TA-DHR2/220623_Rheo/8_3mgml_03%_10to500rads.csv'
file_path4 = '/Volumes/Personal/Student/g16_Yazu/TA-DHR2/220623_Rheo/12_5mgml_03%_10to500rads.csv'
file_path5 = '/Volumes/Personal/Student/g16_Yazu/TA-DHR2/220623_Rheo/25mgml_03%_10to500rads.csv'


df = pd.read_csv(file_path, header=90)
df.drop(index=0, inplace=True)
df2 = pd.read_csv(file_path2, header=90)
df2.drop(index=0, inplace=True)
df3 = pd.read_csv(file_path3, header=90)
df3.drop(index=0, inplace=True)
df4 = pd.read_csv(file_path4, header=90)
df4.drop(index=0, inplace=True)
df5 = pd.read_csv(file_path5, header=90)
df5.drop(index=0, inplace=True)

fig, ax = plt.subplots()

#plt.plot(df['Angular frequency'].astype('float'), df[column].astype('float'), label='2.5mg/mL', marker='.', markersize='6', linestyle='None')
plt.plot(df2['Angular frequency'].astype('float'), df2[column].astype('float'), label='6,25mg/mL', marker='.', markersize='6',linestyle='None', color = "#008000")
plt.plot(df3['Angular frequency'].astype('float'), df3[column].astype('float'), label='8.3mg/mL', marker='.', markersize='6',linestyle='None', color = "#0000cd")
plt.plot(df4['Angular frequency'].astype('float'), df4[column].astype('float'), label='12.5mg/mL', marker='.', markersize='6',linestyle='None', color = "#ff8c00")
plt.plot(df5['Angular frequency'].astype('float'), df5[column].astype('float'), label='25mg/mL', marker='.', markersize='6', linestyle='None', color = "#ff0000")

ax.set_xscale('log')
ax.set_yscale('log')

plt.minorticks_on()
plt.grid(which="major", color="gray", linestyle="solid")
plt.grid(which="minor", color="lightgray", linestyle="dotted")

ax.set_xlabel('Angular frequency[rad/s]')
ax.set_ylabel(column + "[Pa]")

title = column

plt.legend(bbox_to_anchor=(1, 0), loc='lower right', borderaxespad=1)
plt.show()

fig.savefig("%s.png"%title)
```
